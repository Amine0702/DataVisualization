```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 20)
y_cos = np.cos (x)
y_sin = np.sin(x)

plt.plot(x,y_cos,label="cos(x)")
plt.plot(x,y_sin,label="sin(x)")

plt.xlabel("X")
plt.ylabel("Fonctions cosinus et sinus")

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_sin = np.linspace(0, 2*np.pi, 20)
y_sin = np.sin(x_sin)

plt.plot(x_sin,y_sin,"ro-",label="ligne - avec marker = o et couleur=rouge",linewidth=4)
plt.plot(x_sin,y_sin-0.5,"b^--",label="ligne -- avec marker =^ et couleur=bleu")
plt.plot(x_sin,y_sin-1,"g>:",label="ligne : avec marker => et couleur=vert")
plt.plot(x_sin,y_sin-1.5,"c1-.",label="ligne -. avec marker =1 et couleur=cyan")
plt.plot(x_sin,y_sin-2,"mp",label="pas de ligne avec marker =p et couleur=magenta",)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

data = {'Apprentissage supervis':12, 'non supervis':10, 'semi-supervis': 6, 'Autre':16}

Type_algo= list(data.keys())

values = list(data.values())

fig = plt.figure(figsize = (10, 5))


plt.bar (Type_algo, values, color ='maroon', width = 0.4)

plt.xlabel("Type d'algorithme")

plt.ylabel("number d'algorithme")

plt.title("Number of algorithms by type")

plt.show()
```

```python
import matplotlib.pyplot as plt

# set width of bar 
barWidth = 0.25
fig = plt.subplots (figsize = (12, 8))

IT = [12, 30, 1, 8, 22]
ECE = [28, 6, 16, 5, 10] 
CSE = [29, 3, 24, 25, 17]

br1 = np.arange(len(IT)) 
br2 = [x + barWidth for x in br1] 
br3 = [x + barWidth for x in br2]

# Make the plot

plt.bar(br1, IT, color = 'y', width = barWidth, edgecolor = 'grey', label = 'IT') 
plt.bar(br2, ECE, color = 'm', width = barWidth, edgecolor = 'grey', label = 'ECE')

plt.bar(br3, CSE, color = 'r', width = barWidth, edgecolor = 'grey', label = 'CSE')

# Adding Xticks

plt.xlabel('Annee', fontweight = 'bold', fontsize = 10) 
plt.ylabel('Nombre tudiant admis', fontweight = 'bold', fontsize = 10) 
plt.xticks([r + barWidth for r in range(len(IT))], ['2010', '2020', '2021', '2022', '2023'])

plt.legend()

plt.title("Number of students admitted in different years of engineering")

plt.show()
```
